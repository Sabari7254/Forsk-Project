```python
X = data.iloc[:,3:13].values
y = data.iloc[:,13].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('encoder', OneHotEncoder(),[1,2])], remainder='passthrough')
X = np.array(ct.fit_transform(X),dtype=np.float32)
```

```python
X[0:10,:]
```

```python
X = X[:,1:]        #Droping the 1st dummy column for geography
X[0]
```

```python
X = X[:,[0,1,3,4,5,6,7,8,9,10,11]]           #Dropping the column for gender
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
#adding the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation='sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 10)
```

```python
y_pred = classifier.predict_classes(X_test)
#y_pred = (y_pred > 0.5)
```
